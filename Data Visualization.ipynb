# **Data Visualization**
Data visualization is the discipline of trying to understand data by placing it in a visual context so that patterns, trends, and correlations that might not otherwise be detected can be exposed.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
# supervised dataset because has class label
```

# Pandas

## Scatter Plot
To create a scatter plot in Pandas, we can call <dataset>.plot.scatter() and pass it two arguments, the name of the x-column and the name of the y-column. Optionally we can also give it a title.

```python
iris.plot.scatter(x='sepal_length', y='sepal_width', title='Iris Dataset')
```

# Line Chart
To create a line chart in Pandas we can call <dataframe>.plot.line().

```python
# axis=1 is drop column,  axis=0 is row, drop first
# automatically insert legend
iris.drop(['species'], axis=1).plot.line(title='Iris Dataset')
```

# Histogram
In Pandas, we can create a Histogram with the plot.hist method. There arent any required arguments, but we can optionally pass some like the bin size.

```python
# hist put data into bin because it is continuous values, quite messy if plot according to value
# specify how many bin
iris['sepal_length'].plot.hist(bins=20) # default numbers bins = 10
```

Its also straightforward to create multiple histograms.

```python
# subplots : layout=(2,2)->2 rows, 2 columns
#
iris.plot.hist(subplots=True, layout=(2,2), bins=20, figsize=(10,5))
```

# Bar Chart
To plot a bar chart, we can use the plot.bar() method, but before calling this, we need to get our data. We will first count the occurrences using the value_count() method and then sort the occurrences from smallest to largest using the sort_index() method.

```python
# bar chart = value count
iris['petal_length'].value_counts()
```

```python
iris['petal_length'].value_counts().plot.bar()
```

```python
iris['petal_length'].value_counts().sort_index().plot.bar()
```

Its also really simple to make a horizontal bar chart using the plot.barh() method.

```python
# horizontal bar
iris['petal_length'].value_counts().sort_index().plot.barh()
```

# Seaborn
Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for creating attractive graphs.

```python
sns.scatterplot(x=iris['sepal_length'], y=iris['sepal_width'], hue=iris['species'])

plt.title('Iris Dataset')
# hue : color map, use species
```

## Line chart
To create a line chart, the sns.lineplot method can be used. The only required argument is the data, which in our case are the four numeric columns from the Iris dataset.

```python
sns.lineplot(data = iris.drop(['species'], axis=1))
```

# Histogram
To create a histogram in Seaborn, we use the sns.distplot method. We need to pass it the column we want to plot, and it will calculate the occurrences itself. We can also pass it the number of bins and if we want to plot a gaussian kernel density estimate inside the graph.

```python
#sns.displot(iris['sepal_length'], bins=10)
```

```python
sns.distplot(iris['sepal_length'], bins=10)
```

```python
sns.distplot(iris['sepal_length'], bins=10, kde=False) # kde=False if dont want smooth line
```

# Heatmap
A Heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors. Heatmaps are perfect for exploring the correlation of features in a dataset.

To get the correlation of the features inside a dataset, we can call <dataset>.corr(), which is a Pandas dataframe method. This will give us the correlation matrix.

```python
sns.heatmap(iris.corr())
```

```python
# annot=True -> see value density corr
sns.heatmap(iris.corr(), annot=True)
```

# Pairplot
Pairplot enables us to plot a grid of pairwise relationships in a dataset.

```python
sns.pairplot(iris) # remove hue='species'
```

```python
# scatter plot matrix
sns.pairplot(iris, hue='species')
```

# Boxplot
A box plot displays the five-number summary of a set of data. The five-number summary is the minimum, first quartile, median, third quartile, and maximum.

```python
# data = iris dataframe
sns.boxplot(x='species', y='sepal_length', data=iris)
```

```python
sns.boxplot(x='species', y='sepal_length', data=iris, palette='hls')
```

# Matplotlib

## Bar Chart

```python
mean_sl = iris.groupby('species')['sepal_length'].mean()
mean_sl
```

```python
mean_sl.plot(kind='bar')
```

```python
mean_sl.plot(kind='bar', color='gold')
```

```python
mean_sl.plot(kind='bar', color=['gold', 'skyblue', 'violet'])
plt.title('Mean Sepal Length')
plt.xlabel('Species')
plt.ylabel('Mean SL')
plt.xticks(rotation=360)
```
